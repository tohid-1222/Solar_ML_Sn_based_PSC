```python
import pandas as pd
import numpy as np
import pickle
import os
```

```python
pwd = os.getcwd() # current directory
```

```python
df = pd.read_csv(pwd+"/Sn_based/Sn_cleaned.csv") # Import Sn based dataset
encoder = pickle.load(open(pwd+"/Encoded_dataset/TE_enc.pkl","rb")) # Import Target Encoder
```

```python
X_sn = df.iloc[:,1:-4] # Selecting features
y_sn = df.JV_default_PCE # target variable
print(X_sn.shape) # shape of the features
X_sn.head()
```

```python
categorical = list(X_sn.select_dtypes(include="object")) # selecting categorical variables
boolean = list(X_sn.select_dtypes(include="bool")) # selecting boolean variables
```

```python
X_enc = X_sn.copy()
X_enc[boolean] = X_enc[boolean].astype('int') # encoding boolean variables as 0 and 1
X_enc.head()
```

```python
X_enc[categorical] = encoder.transform(X_enc[categorical]) # target encoding categorical columns
X_enc.head()
```

```python
df_enc = pd.concat([X_enc,y_sn],axis=1) # joining encoded features with the target variable 
df_enc.to_csv(pwd+"/Sn_based/df_Sn_enc.csv",index=False) # saving encoded dataset
```
